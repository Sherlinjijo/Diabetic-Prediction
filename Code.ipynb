```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,df.columns!='Outcome'] #data upto 7 rows
y = df.iloc[:,df.columns=='Outcome'] #label outcomes
```

```python
#Train and test we need four variables
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2) #0.2 is 20% of test data and 80% will be test data
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
#To train the data
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain.values.ravel()) #to train the algorithm 
```

```python
predict_output = model.predict(xtest)
print(predict_output)
```

```python
#To check the accuracy of ytest outputs from the xtest input
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(predict_output,ytest)
print('The accuracy score is',acc)
```
